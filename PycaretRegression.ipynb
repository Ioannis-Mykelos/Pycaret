```python
import pycaret
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from pycaret.datasets import get_data
from pycaret.classification import setup
from pycaret.classification import compare_models
from pycaret.classification import tune_model
from pycaret.regression import *
```

```python
#  Import the data
df_reg = get_data('insurance') 
```

```python
# initialize setup
s = setup(df_reg, target = 'charges')
```

```python
# check all the available models
models()
```

```python
# train multiple models
multiple_models = [create_model(i) for i in ['dt', 'lr', 'knn', 'lightgbm']]

# check multiple_models
type(multiple_models), len(multiple_models)

print(multiple_models)
```

```python
# compare all models
best_model = compare_models()
```

```python
# check the best model
print(best_model)
```

```python
# predict on hold-out
pred_holdout = predict_model(best_model)
```

```python
# create copy of data drop target column
data2 = df_reg.copy()
data2.drop('charges', axis=1, inplace=True)

# generate predictions
predictions = predict_model(best_model, data = data2)
```
