```python
import pycaret
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from pycaret.datasets import get_data
from pycaret.classification import setup
from pycaret.classification import compare_models
from pycaret.classification import tune_model
from pycaret.classification import *
```

Load the data for classification


```python
#  Import the data
df_cls = get_data('iris') 
```

# **Classification**:

Iris dataset



```python
s = setup(df_cls, target = 'species')
```

Available Models

```python
# check all the available models
models()
```

Evaluate 1 model that we choose.

```python
model = create_model('lr')
print(model)  #Scikit-Learn object
```

```python
# train multiple models
multiple_models = [create_model(i) for i in ['dt', 'lr', 'knn', 'lightgbm']]

# check multiple_models
type(multiple_models), len(multiple_models)

print(multiple_models)
```

```python
# compare all models
# compare_models returns the output which shows the 10-fold cross-validated metrics for all models.
best_model = compare_models()
```

```python
# check the best model
print(best_model)
```

```python
# predict on hold-out
pred_holdout = predict_model(best_model)
```

To generate predictions on the unseen dataset you can use the same predict_model function but just pass an extra parameter data :

```python
# create copy of data drop target column
data2 = df_cls.copy()
data2.drop('species', axis=1, inplace=True)

# generate predictions
predictions = predict_model(best_model, data = data2)
```
